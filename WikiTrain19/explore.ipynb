# 统计词条出现在摘要第一句中的比例

```python
for language in os.listdir("./full"):
    if "ipynb" not in language:
        count = 0
        point = 0
        print(language)
        for f in os.listdir("./full/"+language+"/"):
            tree = ET.ElementTree(file="./full/"+ language+ "/" + f)
            root = tree.getroot()
            title = root[0].text
            count_summary = len(root[1].text)
            for p in root[1]:
                for sentence in p.text.split(". "):
                    count += 1
                    if title.lower() in sentence.lower():
                        point += 1
                    break
        print(point/count)            
```
